# Importing libraries

```python
import numpy as np
import pandas as pd
```

# Loading Dataset

```python
df = pd.read_csv("Titanic-Dataset.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
#drop duplicates
df.drop_duplicates(inplace=True)
```

```python
df.shape                  #No duplicates existed
```

```python
#find out the null values percentage
df.isnull().sum() / df.shape[0] * 100
```

```python
# Remove columns with more than 5 percent of null values
df.drop(['Age','Cabin'],axis="columns",inplace=True)
```

```python
df.isnull().sum() / df.shape[0] * 100
```

```python
# Discarding unnecessary columns
df.drop(['PassengerId','Name','Ticket','Fare'],axis="columns",inplace=True)
```

```python
df.head()
```

```python
df.drop(['SibSp','Parch','Embarked'],axis="columns",inplace=True)
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

label = le.fit_transform(df['Sex'])

df['Sex'] = label
```

```python
df.head()
```

# Splitting the data


```python
from sklearn.model_selection import train_test_split

X =df.drop('Survived',axis=1)
Y = df['Survived']

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=101)
```

# Model selection - Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()

lr.fit(X_train,Y_train)

y_pred = lr.predict(X_test)


```

# Calculating the accuracy -

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
accuracy = accuracy_score(Y_test,y_pred)
print(f"Accuracy: {accuracy: .2f}")
```

```python
print("Classification report: "+classification_report(Y_test,y_pred))
```
